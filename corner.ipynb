```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread(r'C:\Users\Lenovo\Desktop\myresearch\Ipynb\gosling.jpg')
  
# Setting parameter values
t_lower = 50  # Lower Threshold
t_upper = 150  # Upper threshold
aperture_size = 5 # Aperture size
L2Gradient = True # Boolean
  
# Applying the Canny Edge filter with L2Gradient = True
edge = cv2.Canny(img, t_lower, t_upper, L2gradient = L2Gradient )
  
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
edge = cv2.cvtColor(edge, cv2.COLOR_BGR2RGB)
```

```python
imgs = [img, edge]
title = ['original', 'edge']

plt.subplot(2, 2, 1)
plt.title(title[0])
plt.imshow(imgs[0])
plt.xticks([])
plt.yticks([])


plt.subplot(2, 2, 3)
plt.title(title[1])
plt.imshow(imgs[1])
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
edge = cv2.cvtColor(edge, cv2.COLOR_BGR2RGB)
cv2.imshow('original', img)
cv2.imshow('edge', edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
